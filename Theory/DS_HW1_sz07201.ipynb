## Cleaning Age
To address the issues with the 'Age' column, first, I am replacing incorrect values like '6l' and '7O' with valid numbers ('6' and '70', respectively). Next, convert the column to numeric data types to ensure all entries are valid numbers. Finally, fill any remaining NaN values with the median height.

The **median was used** instead of the mean because the median is less sensitive to outliers and extreme values, which makes it a more robust measure of central tendency for skewed or non-normally distributed data. If the age data includes unusual values or errors, like incorrect entries, the median provides a more accurate central value that is less influenced by these anomalies compared to the mean.
It is generally a safer bet

```python
data_p2['Age (years)'].head()
```

```python
age_clean = data_p2['Age (years)'].str.replace(' ', '', regex=True)
age_clean = pd.to_numeric(age_clean, errors='coerce')
age_clean.isnull().sum() #No null values so no need to put any missing values
```

**Now I am handling the "1993" case here properly.** There is a valid chance that someone might mistakenly put down their year of birth instead of simply putting down their age.

In such scenerios, this helper function detect this and will subtract from the year given from the current year, to find out their real age. 

```python
from datetime import datetime
def calculate_age(value):
    # Check if the value is between 1900 and 2100
    if 1900 <= value <= datetime.now().year:
        current_year = datetime.now().year
        age = current_year - value
        return age
    elif value > datetime.now().year: #This is just in case if someone gives a huge value
        return None
    else:
        return value  # Or handle the case when value is out of range

data_p2['Age (years)'] = age_clean.apply(calculate_age)
data_p2['Age (years)']
```

## Cleaning height

To address the issues with the 'Height (Inches)' column, first, I am replacing incorrect values like '6l' and '7O' with valid numbers ('6' and '70', respectively). Next, convert the column to numeric data types to ensure all entries are valid numbers. Finally, fill any remaining NaN values with the median height.

The **median was used** instead of the mean because the median is less sensitive to outliers and extreme values, which makes it a more robust measure of central tendency for a potentially skewed or non-normally distributed data. 

```python
# Define a function to clean the 'Height (Inches)' column
def clean_height(value):
    try:
        # Attempt to convert value to a numeric type
        value = str(value).replace('O', '0').replace('l', '1')  # Handle common misinterpretations
        return pd.to_numeric(value, errors='coerce')  # Convert to float, coercing errors to NaN
    except:
        return np.nan

# Apply the cleaning function to the 'Height (Inches)' column
data_p2['Height (Inches)'] = data_p2['Height (Inches)'].apply(clean_height)
# Calculate the median
median_height = data_p2['Height (Inches)'].median()

# Fill NaNs with the median and reassign
data_p2['Height (Inches)'] = data_p2['Height (Inches)'].fillna(median_height)
data_p2['Height (Inches)'].head()
```
