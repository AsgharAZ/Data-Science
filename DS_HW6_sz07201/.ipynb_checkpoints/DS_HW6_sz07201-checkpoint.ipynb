## Importing Libraries

```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

#### Unique value columns such as unique id, unique names should not be used in Machine Learning models

```python
df2 = df.drop(['Id'],axis=1)
```

### Question 1 
Create a regression model between GrLivArea and SalePrice (response/output variable). Show the scatter plot with regression line between them. Discuss the R-
squared value

```python
X1 = sm.add_constant(df2['GrLivArea'])
Y1 = df2['SalePrice']
Q1_model = sm.OLS(Y1,X1).fit()
Q1_model.summary()
```

```python
Q1_model.pvalues
```

### Interpretation of the data

R-squared: 0.502 (any value above 0.3 usually shows good relationship between variables) <br>
Y (prediction) = SalePrice <br> X (input) = GrLivArea <br> For each additional unit (feet probably) of above-ground living area on the car, the Sales Price increases by $107

The results indicate a significant positive relationship between above-ground living area (GrLivArea) and sales price (SalePrice) of homes. Specifically, for every additional square foot of living area, the sale price increases by approximately $107.13. The model explains about 50.2% of the variance in sales price (R-squared = 0.502), suggesting a moderate fit. The high F-statistic and its very low p-value (4.52e-223) indicate that the model is statistically significant overall. The high variance does inform us that the ground-living area isn't just a sole indicator of home prices

### Scatter Plot

```python
plt.plot(df2['GrLivArea'].values, df2['SalePrice'].values,'ro')
plt.plot(df2['GrLivArea'].values ,Q1_model.fittedvalues.values,  'b')
plt.legend(['Data', 'Fitted model'])
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
```

On the other hand, while basement area adds value, it's often not as impactful on overall home value or lot area due to building regulations and the perceived importance of above-ground space. ### Multiple Linear Regression (all inputs)

```python
X4 = df2.drop(['SalePrice'],axis=1) #These are all the input variables to "Sale Price" Column
Y4 = df2['SalePrice'] #This is the output "SalesPrice" Column
```

```python
X4.dtypes
```

### Categorical Data Sheningans 
Regression can't work with categorical inputs, One Hot Encoding - For converting categorical columns into numerical columns

```python
one_hot = pd.get_dummies(X4, dtype=int)
one_hot
```

```python
X4 = sm.add_constant(one_hot)
modelcat_Q4 = sm.OLS(Y4, X4).fit()
modelcat_Q4.summary()
```

```python
modelcat_Q4.pvalues
```

### Interpretation of the data

#### Effectiveness of the Model (R-squared Value):
The R-squared value for the model is **0.803**, indicating that 80.3% of the variance in **SalePrice** can be explained by the included features. This is a strong value, suggesting that the model effectively captures the relationship between house features and their sale prices. However, while the R-squared is high, it also suggests that around 19.7% of the variation in **SalePrice** is due to factors not included in the model, implying room for improvement.

#### Three Most Significant Inputs (based on p-value):
1. **OverallQual** (p-value = 7.80e-44): The overall material and finish quality of the house is the most significant predictor of *SalePrice*. A higher quality rating significantly increases the sale price.
2. **GarageCars** (p-value = 6.03e-13): The number of cars that the garage can accommodate is another highly significant factor. Houses with larger garages tend to fetch higher prices.
3. **BsmtFullBath** (p-value = 1.55e-11): The number of full bathrooms in the basement is also a significant factor. Houses with more basement bathrooms tend to have higher sale prices.

#### Three Least Significant Inputs (based on p-value):
1. **HalfBath** (p-value = 0.959): The number of half bathrooms is the least significant factor in predicting sale prices, as indicated by its very high p-value. It appears to have little to no impact on the sale price.
2. **GrLivArea** (p-value = 0.120): Despite being a commonly important feature, the above-ground living area does not appear to significantly affect sale prices in this model.
3. **2ndFlrSF** (p-value = 0.199): The square footage of the second floor also has a relatively high p-value, suggesting it does not play a significant role in predicting the sale price.

The results make sense because features like **Overall Quality**, **Garage Capacity**, and **Basement Full Baths** directly reflect the functional utility and desirability of a home, which strongly influence its market value. A well-constructed home with high-quality materials, ample parking space, and more full bathrooms adds significant value, as buyers tend to prioritize these practical aspects. On the other hand, factors like **Half Bathrooms**, **GrLivArea**, and **Second Floor Square Footage** are less impactful in this model, likely because they contribute less to overall functionality or appeal. For instance, half bathrooms are often seen as secondary amenities, and large living areas or second-floor spaces may not always correspond to buyer preferences as strongly as other features like quality and parking space.

# Question 5 (Predicting LotArea)
Create a regression model to predict LotArea using all other inputs. Discuss the effectiveness of the model using R-squared value. Report 3 most significant inputs and 3 least significant inputs (based on p-value) and interpret the results. Create one new input of your choice of values and show the prediction of LotArea using the same model.  

```python
X5 = df2.drop(['LotArea'],axis=1) #These are all the input variables to "Lot Area" Column
Y5 = df2['LotArea'] #This is the output "SalesPrice" Column
#We already know that X5 will contain some categorical data so doing hot-encodingmodelcat_Q4.summary()
one_hot_2 = pd.get_dummies(X5, dtype=int)
X5 = sm.add_constant(one_hot_2)
modelcat_Q5 = sm.OLS(Y5, X5).fit() #Creating a model
modelcat_Q5.summary()
```

```python
modelcat_Q5.pvalues
```

## Interpretating the results
#### Model Effectiveness Based on R-squared Value:
The R-squared value of the model is 0.172, meaning that the model explains 17.2% of the variance in the LotArea. This indicates that the model has a weak predictive power, suggesting that many important factors affecting the lot area are not captured by the model. An R-squared value this low implies that the model's fit is poor and that other unaccounted variables may be influencing the outcome.

#### Most Significant Inputs (Lowest p-values):
1. **OverallQual (p-value: 4.45e-09)**: This very low p-value shows that the overall quality of the house has a significant negative relationship with *LotArea*. This makes sense because higher-quality houses may be built on smaller, more expensive lots, especially in premium locations.
   
2. **Fireplaces (p-value: 1.13e-06)**: The number of fireplaces is highly significant and positively related to *LotArea*. Homes with more fireplaces are often larger or luxury homes, which typically sit on larger lots.

3. **SalePrice (p-value: 1.36e-05)**: The price of the house is another strong predictor of lot area, which is logical since more expensive properties often include larger lots.

#### Least Significant Inputs (Highest p-values):
1. **OverallCond (p-value: 0.744)**: The overall condition of the house is not a significant predictor of lot area. This makes sense, as the condition reflects the house's upkeep, not the land size.

2. **GrLivArea (p-value: 0.735)**: The above-ground living area has a very high p-value, indicating it is not a significant factor. This is reasonable because a house’s size can vary independently of the lot size, especially in densely built areas where homes are large but sit on smaller lots.

3. **HalfBath (p-value: 0.447)**: The number of half baths does not significantly influence lot size, likely because bathrooms are more about interior features rather than the size of the property.

#### Conclusion:
The regression analysis shows that the most significant predictors of LotArea are OverallQual, Fireplaces, and SalePrice, which make sense because higher-quality homes and those with luxurious features, like fireplaces, tend to be on larger lots. On the other hand, factors like OverallCond, GrLivArea, and HalfBath do not significantly impact lot size, as these aspects focus more on the home's interior and condition rather than the land size. This aligns with common real estate trends, where the quality and luxury of a property, along with its market price, are better indicators of its lot size than the home's internal features.

#### (Evaluating Model Performance: Impact of Feature Selection on SalePrice Prediction)
From Question 4, drop/remove all the columns which are not signification (p-value %> 0.05) and create a new model to predict SalePrice. Discuss the performance of the model using few inputs as compared to using all inputs in (Question 4). Which model do you prefer and why? 

```python
modelcat_Q4.pvalues > 0.05
```

### Observation
As can be observed, variables with p-values above 0.05 are 1stFlrSF, 2ndFlrSF, GrLivArea, BsmtHalfBath, FullBath, and HalfBath so they will be dropped.

```python
Insignificant_Cols = ['1stFlrSF', '2ndFlrSF', 'GrLivArea', 'BsmtHalfBath', 'FullBath', 'HalfBath']
X6 = X4.drop(columns=Insignificant_Cols, axis=1)
X6.head(4)
```

```python
modelcat_Q4_new = sm.OLS(Y4,X6).fit()
modelcat_Q4_new.summary()
```

### Interpreting the results
The performance of this new model shows an R-squared of 0.753, indicating it explains about 75.3% of the variance in SalePrice, which is lower than the old model's R-squared of 0.803. Although the new model captures less variance, it simplifies the analysis by using six fewer features, bringing the total down to 17.

While the new model still maintains significant coefficients for the included predictors, such as OverallQual, **its reduced complexity enhances interpretability**. This is particularly important because a **model with fewer variables is generally easier to communicate to stakeholders and less prone to overfitting**. Additionally, the potential multicollinearity issues present in the old model are likely mitigated in the new model due to the removal of insignificant predictors. In this context, the concept of regularization in regression becomes relevant. Regularization techniques, such as Lasso or Ridge regression, help manage overfitting by penalizing the inclusion of unnecessary predictors. By eliminating insignificant variables in my new model, **I essentially applied a form of regularization, improving the model's interpretability and potentially its generalization to new data**. This approach also helps mitigate multicollinearity issues, which can arise when many predictors are included, further complicating the model.

**Ultimately, I prefer the new model despite its slightly lower R-squared value. Its focus on significant predictors and simplicity make it more actionable for decision-making. I believe that in practical applications, a model that is easier to interpret and communicate is more beneficial, especially when it still retains a good level of predictive power for SalePrice.****
