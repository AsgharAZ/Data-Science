## Data Cleaning

### Skill Ball Control

```python
df['skill_ball_control'].head(10)
```

In this particular column, I simply need to compute the addition or subtraction. For which I can use the split functionality in Python.

```python
def process_value(num):
    if '+' in num:
        base, addition = num.split('+')
        return int(base) + int(addition)
    elif '-' in num:
        base, subtraction = num.split('-')
        return int(base) - int(subtraction)
    else:
        return int(num)
```

```python
df['skill_ball_control'] = df['skill_ball_control'].apply(process_value)

df['skill_ball_control'] = pd.to_numeric(df['skill_ball_control'])
```

Let's check the column again.

```python
df['skill_ball_control'].head(10)
```

It is evident that now all the data has been processed.

### Power Stamina

```python
df['power_stamina'].head(10)
```

We will employ the same strategy as well - using the same function that we defined earlier.

```python
df['power_stamina'] = df['power_stamina'].apply(process_value)

df['power_stamina'] = pd.to_numeric(df['power_stamina'])
```

```python
df['power_stamina'].head(10)
```

This column has also been processed now.

## Generating Summary

### Age Distribution by Nationality - Univariate

```python
age_by_nationality = df.groupby('nationality')['age'].mean().reset_index().sort_values(by='age', ascending=False)

print("Average Age of Players by Nationality (Oldest):")
print(age_by_nationality.head(5))

print("Average Age of Players by Nationality (Youngest):")
print(age_by_nationality.sort_values(by='age', ascending=True).head(5))
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(x='age', y='nationality', data=age_by_nationality[:10], palette='viridis')

plt.xlabel('Average Age')
plt.ylabel('Nationality')
plt.title('Average Age of Players by Nationality (Oldest)')

plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(x='age', y='nationality', data=age_by_nationality[::-10], palette='viridis')

plt.xlabel('Average Age')
plt.ylabel('Nationality')
plt.title('Average Age of Players by Nationality (Youngest)')

plt.show()
```

The results indicate that Grenada, Malta, and Bahrain have the oldest average player ages, suggesting that these countries may rely more on experienced, veteran players. In contrast, Turkmenistan, Sudan, and China PR have the youngest average ages, highlighting a focus on younger talent and potential development. These age dynamics can influence team strategies and player development pathways across different nations.

### Average Overall Rating by Preferred Foot - Bivariate

```python
rating_by_foot = df.groupby('preferred_foot')['overall'].mean().reset_index().sort_values(by='overall', ascending=False)

print("Average Overall Rating by Preferred Foot:")
print(rating_by_foot)
```

The results show that players with a preferred left foot have a slightly higher average overall rating (64.37) compared to right-footed players (64.00). This small difference suggests that left-footed players might be marginally more skilled or valuable in the dataset, possibly due to their relative rarity and strategic advantage in certain positions on the field. However, the difference is minimal, indicating that both left  and right-footed players generally perform at similar levels overall.

```python
summary_table_sorted
```

The summary table reveals that average overall player ratings generally increase with age, reaching a peak around ages 27 to 29, where ratings average between 67.4 and 67.5. This trend suggests that players tend to improve in skill and performance as they gain more experience and develop their abilities. However, after this peak, ratings show a slight decline or stabilization from ages 30 to 41, indicating that while players maintain high performance, the rate of improvement diminishes. The range of ratings, which starts broad for younger players (40 to 75) and narrows slightly for older players, underscores the variability in performance across different ages. Overall, this data highlights that players typically reach their peak performance in their late 20s to early 30s, with a gradual stabilization or decline as they age further.

### Pace vs Shooting (Continuous vs Continuous)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='dribbling', y='skill_ball_control', alpha=0.5, color='green')
plt.xlabel('Dribbling')
plt.ylabel('Skill Ball Control')
plt.title('Dribbling vs. Skill Ball Control')
plt.show()
```

The scatter plot of dribbling versus skill ball control shows a nearly linear relationship, indicating a strong positive correlation between the two attributes. As dribbling skills increase, players tend to have higher skill ball control. This linear trend suggests that players who excel at dribbling are likely to possess better ball control, reinforcing the idea that these two skills are closely linked. The consistent upward trend in the plot implies that improvements in dribbling ability are generally accompanied by enhancements in ball control, highlighting their interdependence. This relationship can be crucial for understanding player development and performance, as it underscores the importance of refining dribbling skills to achieve superior ball control.

We can confirm this relationship by generating a heatmap.

```python
plt.figure(figsize=(12, 10))
correlation_matrix = df[['dribbling', 'skill_ball_control']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Dribbling and Skill Ball Control')
plt.show()
```

A correlation coefficient of 0.89 is quite high, suggesting a robust positive relationship between dribbling and skill ball control. Hence, while developing training programs, improvements in one attribute (like dribbling) are likely to be associated with improvements in the correlated attribute (like skill ball control). 

### Overall Rating vs Work Rate (Continuous vs Discrete)

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df, x='work_rate', y='overall', palette='Set1')
plt.xlabel('Work Rate')
plt.ylabel('Overall Rating')
plt.title('Overall Rating by Work Rate')
plt.xticks(rotation=90)
plt.show()
```

The analysis of the relationship between `work_rate` and overall rating reveals several key insights. Players with a high work rate in both offensive and defensive roles (`high/high`) tend to have the highest average overall ratings. This suggests that a strong work ethic in both dimensions is associated with superior performance, highlighting the value of tireless effort in contributing effectively to both attacking and defensive plays. In contrast, players with a medium work rate in both areas (`medium/medium`) show the lowest lower quartile ratings, indicating that a medium work rate might be less advantageous, leading to lower overall effectiveness compared to those with higher work rates. Players with low work rates in both aspects (`low/low`) exhibit the smallest range in overall ratings, implying a more consistent but generally lower performance level. Interestingly, players with a medium offensive work rate and a low defensive work rate (`medium/low`) present the highest rating outliers, demonstrating that even with a lower defensive work rate, their performance can be exceptional due to their strong offensive contribution. Overall, this analysis highlights that a high work rate, particularly in both roles, generally correlates with better performance, although some players excel despite having a mixed work rate profile.

### Correlation of Player Attributes (Correlation)

```python
plt.figure(figsize=(12, 10))
correlation_matrix = df[['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic', 'skill_ball_control', 'power_stamina']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Player Attributes')
plt.show()
```

The correlation analysis reveals several notable relationships among the attributes related to player performance. There is a very high positive correlation of 0.89 between skill ball control and dribbling, indicating that players who excel in one are likely to excel in the other. Similarly, passing and dribbling have a strong positive correlation of 0.8, while dribbling and shooting also show a significant positive correlation of 0.76, suggesting that players skilled in dribbling are often proficient in passing and shooting as well. The correlation between dribbling and pace is 0.57, highlighting a moderate positive relationship where faster players tend to be better dribblers.

On the other hand, power stamina shows a generally low positive correlation with other attributes, with the highest correlation being 0.53 with physic, indicating that while there is a positive relationship, it is relatively weak compared to other correlations. Defending and shooting exhibit the strongest negative correlation at -0.45, suggesting that players who are better at defending may not be as effective in shooting. Additionally, all attributes related to pace, shooting, and dribbling show negative correlations with physic and defending, implying that as players focus more on these attributes, their performance in defending and physicality may decrease. Overall, defending appears to be the least positively correlated with most fields, highlighting its distinct role compared to other attributes.
