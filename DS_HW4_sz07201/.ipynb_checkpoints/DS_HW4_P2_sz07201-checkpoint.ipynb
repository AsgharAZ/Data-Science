## Bivariate Analysis
### Relationship Between Land Area and Population

```python
plt.figure(figsize=(6, 3))
sns.scatterplot(x='country_area_land', y='country_population', data=df_countries)
plt.title('Country Area vs. Population')
plt.xlabel('Land Area (sq km)')
plt.ylabel('Population')
plt.show()
```

```python
plt.figure(figsize=(6, 2))
correlation_matrix = df_countries[['country_area_land', 'country_population']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix between Country_Area_Land and Country_Population')
plt.show()
```

As we can infer from the Scatter Plot as well as additionally confirm our suspicions by making a correlation Matrix. We can confirm that there is almost no correlation between country having a large land mass and having a larger population. 

So it means, even if a country is pretty large, you can't say for sure that it will also have a larger population size.

```python
plt.figure(figsize=(6, 2))
correlation_matrix = df_countries[['country_area_water', 'country_population']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix between Country_Area_Water and Country_Population')
plt.show()
```

A country having a larger water land to itself isn't indicative of it having a large population either. 

## Multivariate Analysis


```python
merged_df = pd.merge(df_countries, df_cities, on='fact_id')
plt.figure(figsize=(12, 3))
correlation_matrix = merged_df[['country_population', 'city_population', 'country_area', 'country_area_land', 'country_area_water','country_birth_rate', 'country_death_rate', 'country_migration_rate']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', fmt='.2f', linewidths=0.5)
plt.show()
```

Identifying some correlations

### Explaining the difference in correlation found before and after merging country and city data-set, especially between population and land area
The shift in correlation between population and land area, from 0.10 (weak positive linear correlation) in the countries dataset to 0.45 (moderate positive linear correlation) after merging with the cities dataset, highlights how data composition affects statistical relationships. Initially, the weak correlation reflects the diversity of countries, where factors like population density and land size vary widely. Larger countries can have low population densities, leading to a diluted correlation.

After merging, the introduction of city data changes this dynamic significantly. Urban areas tend to concentrate populations, meaning that larger countries often contain larger cities with higher populations. This commonality enhances the correlation, as the relationship between land area and population becomes more pronounced when considering urban centers. For example, populous cities like Algiers and Luanda skew the average, revealing a stronger linkage that wasn't as apparent at the country level alone.

Additionally, the merged dataset includes multiple entries for countrieseach corresponding to a citycreating repeated measures. This repetition further strengthens the correlation because larger land areas are likely to host multiple populous cities. This change suggests that policymakers and planners should consider urbanization trends when analyzing demographic data. Understanding these relationships is crucial for effective resource allocation and infrastructure development, especially in countries with significant urban centers.

In summary, the increased correlation emphasizes the need for careful interpretation of statistical metrics when combining datasets from different levels. It illustrates the intricate interplay between geography and population dynamics, highlighting the importance of urbanization in shaping demographic trends.

### Correlation between country_area , counter_area_land, country_area_water
#### Correlation between country_area and country_area_land
The high correlation between "country_area_land" and "country_area" (1.00) is expected because "country_area" represents the total area of a country, which includes both land and water areas. For most countries, the land area ("country_area_land") constitutes the majority of the total area, especially in countries where the water bodies are relatively small or non-existent. In such cases, the "country_area" is almost entirely determined by the "country_area_land", leading to a perfect or near-perfect correlation.

#### Correlation between country_area and country_area_water
The correlation of 0.85 between "country_area" and "country_area_water" can be explained by the fact that countries with larger total areas tend to have larger water bodies as well. This relationship suggests that as a country's overall size increases, it is likely to contain more water bodies, whether lakes, rivers, or coastal areas. However, the correlation is not perfect (less than 1.00), which indicates that there are countries with large land areas but smaller water bodies and vice versa. This makes sense cause if you were to see the world map, the water and land is not at all evenly distributed.

#### Correlation between country_area_land and country_area_water
Similarly, the 0.84 correlation between "country_area_land" and "country_area_water" points to the same trend: countries with more land area also tend to have more water area. However, the correlation being less than 1.00 shows that this is not a consistent rule across all countries. 

These correlations indicate that while land and water areas are related, the extent of their relationship varies based on geographical features, such as the presence of coastlines, lakes, and rivers, which influence the distribution of land and water across countries.0
